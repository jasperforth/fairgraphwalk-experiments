# Plot x=Accuracy y=F1-score Scatter plot

the filepaths should point to the relevant combined results csvs, e.g. new_pokec_mixed_graphsize_small2_0_results_fmmc.csv

```python
import pandas as pd
import matplotlib.pyplot as plt

crosswalk_filepath = ''
baseline_filepath = ''
fmmc_filepath = ''

crosswalk_data = pd.read_csv(crosswalk_filepath)
baseline_data = pd.read_csv(baseline_filepath)
fmmc_data = pd.read_csv(baseline_filepath)

# print(crosswalk_data)
# df = pd.read_csv('/Volumes/tcs_jf_fair_node_sampling/test_exprmt_diff_nr_graphs/data/reports/combined_results/pokec_distinct_0_results_cw_sens_AGE.csv')
# df.plot(kind = 'scatter', x = 'accuracy', y = 'f1_macro', ylabel = 'disparity')

crosswalk_data = crosswalk_data.drop(crosswalk_data[crosswalk_data.sens_attr=='label_region'].index)
# print(crosswalk_data)

ax = crosswalk_data.plot(kind = 'scatter', x = 'accuracy', y = 'f1_macro', ylabel = 'f1-score')
fmmc_data.plot(ax=ax, kind = 'scatter', x = 'accuracy', y = 'f1_macro', ylabel = 'f1-score', color='red')
# plt.scatter(baseline_data['accuracy'], baseline_data['f1_macro'], color='green')
plt.legend(["Crosswalk", "FMMC"])
plt.show()
```
